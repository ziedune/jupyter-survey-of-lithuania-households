```python
# Distribution of persons by region and sex in 2015 and 2019:
def get_count_by_region_and_gender(region, gender, year):
    q = "select count(*) from {}_asmenys where RB090 = '{}' and M_K = '{}' and age >=18".format(
        year, gender, region
    )
    c.execute(q)
    ats = c.fetchall()
    return ats[0][0]


count_townen_in_town15 = get_count_by_region_and_gender(1, 1, 2015)
count_women_in_town15 = get_count_by_region_and_gender(1, 2, 2015)
count_townen_in_countryside15 = get_count_by_region_and_gender(2, 1, 2015)
count_women_in_countryside15 = get_count_by_region_and_gender(2, 2, 2015)
count_townen_in_town19 = get_count_by_region_and_gender(1, 1, 2019)
count_women_in_town19 = get_count_by_region_and_gender(1, 2, 2019)
count_townen_in_countryside19 = get_count_by_region_and_gender(2, 1, 2019)
count_women_in_countryside19 = get_count_by_region_and_gender(2, 2, 2019)
```

```python
# DRAWING-boxes to pie:
def add_boxes_to_pie(axs, labels, wedges):

    bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
    kw = dict(arrowprops=dict(arrowstyle="-"), bbox=bbox_props, zorder=0, va="center")

    for i, p in enumerate(wedges):
        ang = (p.theta2 - p.theta1) / 2.0 + p.theta1
        y = np.sin(np.deg2rad(ang))
        x = np.cos(np.deg2rad(ang))
        horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
        connectionstyle = "angle,angleA=0,angleB={}".format(ang)
        kw["arrowprops"].update({"connectionstyle": connectionstyle})
        axs.annotate(
            labels[i],
            xy=(x, y),
            xytext=(1.35 * np.sign(x), 1.2 * y),
            horizontalalignment=horizontalalignment,
            **kw
        )
```
